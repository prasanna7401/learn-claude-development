```python
# Install dependencies
%pip install anthropic python-dotenv

# Load environment variables from .env file
from dotenv import load_dotenv

load_dotenv()

# Create API client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

**Multi-turn Conversations** 

- LLM cannot store conversations, so we need to setup functions to track the list of messages and share that in every followup request.
- Create helper functions to save those client messages

```python
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text
```

**Process**

1. Create an empty list of `messages[]`   
2. Add initial question of user to this (`message.append()`)
3. Pass the message list to `chat()`
4. Add the response from LLM as an `assistant` message into the list.
5. Add this in every follow-up question of the user.
6. Continue chat whenever required with this conversation history.

```python
# 1 - Initialize an empty list to hold the conversation messages
messages = []

# 2 - Append user question to the messages list
add_user_message(messages, "What is the capital of France?") # Replace with user input in a real application

messages # for demonstration purposes, we can see the messages list before sending it to the API
```

```python
# 3 - Send the conversation messages to the LLM API and get the response
response = chat(messages)

response # for demo purpose
```

```python
# 4 - Append the assistant response to the messages list
add_assistant_message(messages, response)

messages
```

```python
add_user_message(messages, "What is that place famous for?") # User's followup question

response = chat(messages)

print(response)
```
