### Task

1. Prompt the user to enter some input using the built-in input function.
2. Add it to the list of messages.
3. Call the API
4. Add generated text to the list of messages.
5. Print the generated text.
6. Repeat from #1

```python
# Install dependencies
%pip install -q anthropic python-dotenv

# Load environment variables from .env file
from dotenv import load_dotenv

load_dotenv()

# Create API client

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
def add_user_message(messages, text):
    messages.append({"role": "user", "content": text})

def add_assistant_message(messages, text):
    messages.append({"role": "assistant", "content": text})

def chat(messages):
    response = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )
    return response.content[0].text
```

```python
messages = []

while True:
    try:
        user_input = input("Enter your question: ").strip()
        print(f"\nYou: {user_input}")
        
        if not user_input:
            print("Please enter a valid question.")
            continue
        
        add_user_message(messages, user_input)
        response_text = chat(messages)
        add_assistant_message(messages, response_text)
        print(f"\nAssistant: {response_text}\n")
        
    except (KeyboardInterrupt, EOFError):
        print("\nExiting chat. Goodbye!")
        break
```
