# System Prompts

**Using System Prompts create a specialized chatbot**

Role: Math Tutor Specialist

- DOs:
    - Initially only give hints
    - Patiently guide the student towards the solution
    - Provide example solutions for similar problems

- DO NOTs:
    - Immediately give the answer
    - Ask the student to cheat or use calculator

```python
# Install dependencies
%pip install -q anthropic python-dotenv

from anthropic import Anthropic

# Load environment variables from .env file
from dotenv import load_dotenv

load_dotenv()

# Create API client
client = Anthropic()
model = "claude-sonnet-4-6"

# Converation History Helper Functions
def add_user_message(messages, text):
    messages.append({"role": "user", "content": text})

def add_assistant_message(messages, text):
    messages.append({"role": "assistant", "content": text})
```

**System Prompt**

- A system prompt is the initial instruction that defines the assistant’s role, behavior, and response boundaries for the conversation.

> Note: 
> - Here, we are defining the system prompt outside of the chat() function for customizability and not hardcoding it.
> - So, we pass the input data as **params**
> - **Why `**params`**?
>       Gives the ability to handle when no system prompt is provided (`None`), and added only when system prompt is provided, rather than directly passing `system=None` to LLM API.


```python
def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }

    if system:
        # If a system prompt is provided, include it in the parameters
        params["system"] = system

    response = client.messages.create(**params)
    return response.content[0].text


# Main Program Flow

messages = []

system_prompt = '''
You are a Math Tutor Specialist. Your job is to help students understand how to solve mathematical problems.

#### DOs:
    - Initially only give hints
    - Patiently guide the student towards the solution
    - Provide example solutions for similar problems

#### DO NOTs:
    - Answer to queries outside of your role (outside mathematics)
    - Immediately give the answer
    - Ask the student to cheat or use calculator

'''

while True:
    try:
        user_input = input("Enter your question: ").strip()
        print(f"\nYou: {user_input}")
        
        if not user_input:
            print("Please enter a valid question.")
            continue
        
        add_user_message(messages, user_input)
        response_text = chat(messages, system=system_prompt)
        add_assistant_message(messages, response_text)
        print(f"\nAI Tutor: {response_text}\n")
        print("-" * 50)  # Separator for readability
        
    except (KeyboardInterrupt, EOFError):
        print("\nExiting chat. Goodbye!")
        break
```
